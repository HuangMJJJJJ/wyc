```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(19, 10), dpi=10)
df = pd.read_csv("result_dif_range_num_nonmat_w2sim.csv")
ax = sns.lineplot(
    x="train ratio",
    y="base acc ratio",
    data=df,
)
ax = sns.lineplot(
    x="train ratio",
    y="methon1 acc ratio",
    data=df,
)
ax = sns.lineplot(
    x="train ratio",
    y="methon2 acc ratio",
    data=df,
)
ax = sns.lineplot(
    x="train ratio",
    y="methon3 acc ratio",
    data=df,
)
ax.legend(
    ["base acc ratio", "methon1 acc ratio", "methon2 acc ratio", "methon3 acc ratio"]
)
ax.set_ylim(0.8, 1)
# plt.legend()
plt.show()
```
